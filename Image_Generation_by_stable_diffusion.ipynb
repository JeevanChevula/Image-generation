```python
from diffusers import StableDiffusionPipeline
import torch

# Load the Stable Diffusion model
model_id = "stabilityai/stable-diffusion-2"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

# Generate an image from a textual description
prompt = "An ancient floating city suspended in the sky by massive chains anchored to distant mountain peaks. The architecture blends Gothic spires with glowing bioluminescent crystals. Airships glide between towers under a twilight sky filled with stars and nebulae. Waterfalls cascade from the city's edges into the clouds below. Magical runes glow faintly along stone bridges, and hooded figures in flowing robes walk along illuminated pathways"
image = pipe(prompt).images[0]

# Save the generated image
from google.colab import drive
drive.mount('/content/drive')
image.save("/content/drive/My Drive/Colab Notebooks/Generated Images/Ancient city image.png")
print("Image saved successfully!")
```

```python
from diffusers import StableDiffusionPipeline
import torch

# Load the Stable Diffusion model
model_id = "stabilityai/stable-diffusion-2"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

# Generate an image from a textual description
prompt = ["A rainy scene where a flooded lake overflows onto a nearby road. The water covers the road up to 10 cm, creating a shallow stream across it. A crocodile and fish are swimming through the water on the road. The environment is realistic, with wet reflections, dark clouds, and a natural atmosphere. Photorealistic style, high detail, 4K resolution",
          "A serene lake beside a winding road, after a heavy downpour, water overflows from the lake and floods the road, with a crocodile and fish swimming in the 10cm-deep water, amidst a misty atmosphere"]
images = pipe(prompt).images

# Save the generated image
from google.colab import drive
drive.mount('/content/drive')
images[0].save("/content/drive/My Drive/Colab Notebooks/Generated Images/Image_with_chatgpt_prompt_normalsettings.png")
images[1].save("/content/drive/My Drive/Colab Notebooks/Generated Images/Image_with_llama_prompt_normalsettings.png")
print("Images saved successfully!")
```
